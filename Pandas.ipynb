# **Series Changing Index**

```python
ball=["pink", "white", "black", "blue"]
occurence = [20, 15, 6, 43]
S=pd.Series(occurence, index=ball)
print(S)
```

# **How to Create DataFrames**

1-From a list:

```python
list = [['Jack', 34, 'Paris'], ['Thomas', 30, 'Roma'],
       ['Alexandre', 16, 'New York']]
df = pd.DataFrame(list)
df
```

```python
list = [['Jack', 34, 'Paris'], ['Thomas', 30, 'Roma'],
       ['Alexandre', 16, 'New York']]
df = pd.DataFrame(list, columns =['name', 'age', 'city'])
df
```

```python
df = pd.DataFrame(list, index =['name', 'age', 'city'])
df
```

```python
df = pd.DataFrame(list, ['name', 'age', 'city'])
df
```

2-From a dictionary:

```python
dictionary  = { 'name' : ['Jack', 'Thomas', 'Alexandre'],
    'age' : [34, 30, 16],
    'city' : ['Paris', 'Roma', 'New York']}

df = pd.DataFrame(dictionary, columns =['city'])
print(df)
```

3-From a NumPy array:

```python
import numpy as np
my_numpy_array=np.random.randn(3,4)
my_numpy_array
```

Here's what .randn(3,4) means:

.randn(): This function returns samples from the standard normal distribution. Unlike other distributions, there are no parameters needed to specify the mean and standard deviation since they are fixed at 0 and 1, respectively.

3,4: These are the dimensions of the array that .randn() will generate. 3 represents the number of rows and 4 represents the number of columns.

 So, .randn(3,4) will produce an array with 3 rows and 4 columns.

# **Getting Information about DataFrame**

```python
df=pd.read_csv("iris.csv")
df
```

df.head()

Purpose: To return the first n rows of the DataFrame.

Usage: df.head(n) where n is the number of rows you want to see. If you don't specify n, it defaults to 5.

Common Scenario: Use this when you want to quickly check the structure and content of the DataFrame's beginning, such as seeing column names and some initial values.

```python
df.head()
```

df.tail()

Purpose: To return the last n rows of the DataFrame.

Usage: df.tail(n) where n is the number of rows you want to see. If you don't specify n, it defaults to 5.

Common Scenario: Use this to quickly check the end of your dataset, which can be especially useful to confirm the data's order or the outcome of recent manipulations.

```python
df.tail()
```
